```python
movies.iloc[0].genres
# Here data is in the form of list of dic or string of list 
```

```python
# Extract the name from string of list
import ast 
def convert(obj): 
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```
