```python
# Extract the name from string of list
import ast 
def convert(obj): 
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    
```

```python
# Extract the name from string of list
import ast 
def convert3(obj): 
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3: 
            l.append(i['name'])
            counter+=1
        else: 
            break
    return l
    
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj): 
    l=[]
    for i in ast.literal_eval(obj): 
        if i['job']=='Director': 
            l.append(i['name'])
            break
    return l
```

## Bag Of Words(BoW)
* Bag of Words (BoW) is a technique in Natural Language Processing (NLP). 
* It is widely used to transform textual data into machine-readable format.

```python
# Now Its time for Text vecterization 
# We calculating similarity score between the movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english') 
```

```python
# A sparse matrix is a matrix where the majority of elements are zero, 
# typically containing only a small fraction of non-zero values.
```

```python
Text_vector=cv.fit_transform(new_df['tags']).toarray()
```

```python
Text_vector
```

```python
print(cv.get_feature_names_out())
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split(): 
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
ps.stem('loved','loving')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(Text_vector)
```

```python
def recommended(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list: 
        print(new_df.iloc[i[0]].title)
        
    
    
```
